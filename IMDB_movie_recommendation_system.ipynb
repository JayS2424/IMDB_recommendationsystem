# Preprocessing

## Converting Each Element of column into list

```python
import ast
from ast import literal_eval
literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def extract_genres(text):
    genres=[]
    for i in ast.literal_eval(text):
        genres.append(i['name'])
    return genres
        
```

```python
tmdb['genres']=tmdb['genres'].apply(extract_genres)
```

```python
tmdb
```

```python
tmdb['keywords']=tmdb['keywords'].apply(extract_genres)
```

```python
tmdb
```

```python
tmdb['cast'][0]
```

```python
def cast_extracter(text):
    cast=[]
    counter=0
    for i in ast.literal_eval(text):
        if counter!=3:
            cast.append(i['name'])
            counter+=1
        else:
            break
    return cast
        
```

```python
tmdb['cast']=tmdb['cast'].apply(cast_extracter)
```

```python
tmdb
```

```python
tmdb['crew'][0]
```

```python
def crew_director(text):
    crew=[]
    for i in ast.literal_eval(text):
        if i['job']=='Director':
            crew.append(i['name'])
            break
    return crew
```

```python
tmdb['crew']=tmdb['crew'].apply(crew_director)
```

```python
tmdb['overview']=tmdb['overview'].apply(lambda x:x.split())
```

```python
tmdb.head()
```

```python
tmdb['genres']=tmdb['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
tmdb['keywords']=tmdb['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
tmdb['cast']=tmdb['cast'].apply(lambda x:[i.replace(" ","") for i in x])
tmdb['crew']=tmdb['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
tmdb.head()
```

## Merging all columns into one column 'tags'

```python
tmdb['tags']=tmdb['overview']+tmdb['genres']+tmdb['keywords']+tmdb['cast']+tmdb['crew']
```

```python
tmdb.head()
```

```python
new_df=tmdb[['movie_id','title','tags']]
```

```python
new_df["tags"][0]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'][0]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

# Stemming all words in the tags column

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    
    return " ".join(y)   
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

# Vectorization of the Stemmed Words

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=10000,stop_words='english')
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.vocabulary_
```

# Calculation of the Cosine Similarity for recommondation

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
sorted(similarity[0],reverse=True)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
new_df[new_df['title']=='Avatar'].index[0]
```

# Function For reccomendation

```python
def reccomend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distances=similarity[movie_index]
    movie_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
        
        
```

```python
reccomend('Avatar')
```

```python
reccomend('John Carter')
```
